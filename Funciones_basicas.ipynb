```python
#funciones
#def Nombre_funcion(parametro): esta es la sintaxis de una función
#    intrucciones
#    return(opcional)

def prueba():
    print("Hola mundo")

    
def suma():
    num1=5
    num2=7
    print(num1 + num2)

def suma2(n1, n2):
    print(n1 + n2)

def suma3(n3, n4):
    res = n3 + n4
    return res
#Si la función regresa un valor (return) 
#no lo va a mostrar cuando sea llamada hasta que se imprima (print)

# Veremos ahora las listas (vectores)
# las listas en python permiten guardar distintos tipos
# de valores, no así en otros lenguajes como java
# python además permite expandir la lista dinámicamente
# es decir se puenden agregar valores a la lista de 
# manera continua como en matlab

#Sintaxis de la lista
# nombre_lista=[elem1,elem2,elem3,...]

# indice es la posición del elemento dentro de la lista
# se comienza a numerar los elementos desde 0
# ej lista=[15,30,22,8] si queremos llamar al primer
# elemento de la lista (15) hacemos lista[0]=15

lista=["María", "Pepe", "Martha", "Antonio"]

#para agregar un elemento al final de la lista utilizamos el comando .append
lista.append("Simon")

#para agregar un elemento en un punto intermedio de la lista utilizamos la funcion .insert
lista.insert(2,"Julio")  #primer elemento de la funcion insert es el indice de la lista en donde queremos
#agregar el nuevo elemento

#Para extender la lista o concatenar (unir dos listas) utilizamos
# la funcion .extend
lista.extend(["Pablo", "Armando", "Luz"])

#si queremos saber el indice dentro de una lista de un valor en particular
#utilizamos la funcion .index
print(lista.index("Simon"))

#si queremos saber si un valor en particular se encuentra o no
#dentro de la lista utilizamos la funcion in, esto nos devuelve un valor 
#booleano true o false
print("Pepe" in lista)

#Ejemplo de lista con distintos valores (struc, int, boolean, float)
lista2 = ["Simon", 521, True, 5.29]

#si queremos eliminar un elemento en particular de la lista utilizamos
#la funcion remove
lista2.remove("Simon")
print(lista2[:])

#Si queremos eliminar el ultimo elemento de la lsita se utiliza la
#funcion .pop
lista2.pop()
print(lista2[:])
#Podemos sumar dos listas con el operador suma
lista3=lista+lista2
print(lista3[:])
#Podemos repetir una lsita con el operador multiplicación
lista4=["Ana",3.14,True]*3
print(lista4[:])


```

```python
prueba()
suma()
suma2(13,25)
print(suma3(13,30))

print(lista[:])     #esto permite imprimir todos los valores de la lista
print(lista[0])
print(lista[-2])    #esto permite imprimir un valor comenzando desde el final
print(lista[0:2])   #imprime los tres primeros 
print(lista[:2])    #imprime todos los valores del indice que van desde 0 hasta 2
print(lista[2:])    #imprime todos los valores del indice que van desde el 2 hasta el final

```

```python
#----------------------------------------------------------------
#Tema: Las tuplas
#Son similares a las listas pero con ciertas restricciones
#Las tuplas no pueden ser modificadas (no añaden objetos, no se concatenan con otras listas,
#no se remueven objetos, etc.) 
#Las tuplas permiten identificar si un objeto existe o no dentro de la tupla
#Las tuplas permiten extraer elementos de ellas pero su resultado es otra tupla
#Ventajas:
#Son más rapidas, ocupan menos espacio en memoria(optimización), formatean strings(por verse)

#sintaxis de las tuplas
#NombreTupla=(elem1,elem2,elem3,...) 
#a diferencia de las listas, las tuplas van entre parentesis

tupla=("simon",2,3.14,True)
print(tupla[1])
lista=list(tupla)
print(lista)
print(tupla)
tupla2=tuple(lista)
print(tupla2)
print("simon" in tupla)
#funcion count permite contar la cantidad de elementos existentes en la tupla
print(tupla.count("simon"))
#funcion len te dice cuantos elementos hay en la tupla
print(len(tupla))
#fn+supr activa/desactiva la funcion overwrite
#desempaquetdo de tuplas (guardar en variables los valores dentro de la tupla)
nombre, dia, pi, bol = tupla
print(nombre)
print(dia)
print(bol)
print(pi)
```

```python
#-------------------------------------------------------------------------
#Tema: Los diccionarios
#Estructuras de datos parecidas a las listas y tuplas
#principal caracteristica de los diccionarios es que los datos se almacenan 
#asociados a una clave de tal forma que se crea una asociacion de tipo
#clave:valor para cada elemento almacenado
#a cada valor se le asigna una clave particular
#En los diccionarios podemos almacenar listas, tuplas y otros diccionarios
#Los elementos almacenados no están ordenados. el orden es indiferente 
#a la hora de almacenar información en un diccionario
diccionario={"Alemania":"Berlín","Francia":"París","Reino Unido":"Londres", "España":"Madrid"} 
print(diccionario["Francia"])
diccionario["Italia"]="Lisboa"
print(diccionario)
#Los valores que vayamos agregando a una clave en particular se van sobreescribiendo sobre
#los valores anteriores
diccionario["Italia"]="Roma"
print(diccionario)
#para eliminar una pareja de elementos(clave:valor) se utiliza la funcion del
del diccionario["Reino Unido"]
print(diccionario)

#también es posible hacer combinaciones de string-valor numerico
diccionario2 = {"Alemania":"Berlín", 23:"Jordan", "Mosqueteros":3}
print(diccionario2)

#Podemos crear un diccionario a partir de elementos de una tupla o lista
tup=["España", "Francia", "Reino Unido", "Alemania"]
diccionario3 = {tup[0]:"Madrid",tup[1]:"París",tup[2]:"Berlin"}
print(diccionario3)
print(diccionario["Francia"])

#Podemos ingresar listas o tuplas dentro de un diccionario
tup2={23:"Jordan", "Nombre":"Michael","Equipo":"Chicago","Anillos":[1991,1992,1993,1996,1997,1998]}
print(tup2)
print(tup2["Equipo"])
print(tup2["Anillos"])

tup3={23:"Jordan", "Nombre":"Michael","Equipo":"Chicago","Anillos":{"Temporadas":[1991,1992,1993,1996,1997,1998]}}
print(tup3["Anillos"])

print(tup3.keys())
print(tup3.values())
```
